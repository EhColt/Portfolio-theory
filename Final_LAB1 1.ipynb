## Task #5 : Calculate stock returns (1)

```python
# daily_return as a function to calculate stocks daily returns
# we used another way to find the stocks daily returns instead of using loops
def daily_return(data):
    prices = data.iloc[:,1:].reset_index()
    # with p1 we take the section of the dataframe from second row to the end 
    p1 = prices.iloc[1:,:].reset_index().drop(['level_0','index'],axis=1)
    # with p0 we take the section of the dataframe from the first row to the penultimate row
    p0 = prices.iloc[:-1,:].reset_index().drop(['level_0','index'],axis=1)
    # In this way we can subtract each row (starting from the second row) with the previous one in order to obtain the daily changes of the stock prices
    p1_p0 = p1-p0
    # And so we can obtain the daily returns using the daily changes
    returns = p1_p0/p0
    returns['Date'] = data['Date'][:-1].reset_index().drop(['index'],axis=1)
    date = returns.pop('Date')
    returns.insert(0,'Date',date)
    return returns
plot_stocks(daily_return(data),'Daily return')
```

## Task #6 : Portfolio weights


**Simulate 1000 portfolios with random weights**

```python
np.random.seed(2002) 
weights = np.random.rand(1000,8) 
weights = weights/weights.sum(axis=1).reshape(-1,1)
weights
```

```python
np.dot(weights[0],d_r.iloc[:,1:-1].mean().values)*252
```

```python
# The average return of each portfolio
def weighted_return(d_r,weights):
    return np.dot(weights,d_r.iloc[:,1:-1].mean().values)*252 #number of trading days in a year
w_r = weighted_return(d_r,weights)
w_r
```

```python
# Standard deviation of each portfolio 
def weighted_std(d_r,weight):
    return np.sqrt(weight @ (d_r.iloc[:,1:-1].cov()*252) @ weight.T)
w_std = []
for weight in weights:
    w_std.append(weighted_std(d_r,weight))
w_std = np.array(w_std)
w_std
```

```python
# Sharpe ratio of each portfolio
rf = 0
sharp_ratio = (w_r-rf)/w_std # Sharpe ratio
sharp_ratio
```

```python
# The best portfolio with the highest sharpe ratio 
weights[np.where(sharp_ratio==np.max(sharp_ratio))]
```

```python
#Plot of the 1000 random portfolio 
fig = plt.figure(figsize=(12,6),dpi=300)
plt.scatter(w_std,w_r,c=sharp_ratio,cmap='Blues')
plt.scatter(w_std[np.where(sharp_ratio==np.max(sharp_ratio))],w_r[np.where(sharp_ratio==np.max(sharp_ratio))],s=50,c='red',label='tangency portfolio')
plt.xlabel('volatility')
plt.ylabel('return')
plt.show()
```

```python
d_r_temp = d_r.copy() # a copy of daily returns to avoid the re-run of all previuos codes 
d_r_temp.insert(9,'rf',[rf/253]*d_r.shape[0]) # insert the risk free into the dateframe 
```

```python
# The efficient frontier
def eff_frontier(return_level,d_r):
    f = lambda x: weighted_std(d_r,x)
    bounds = tuple([(None,None) for i in range(d_r.shape[1]-2)])
    res = []
    w = []
    for r in return_level:
        cons = (
            {'type':'eq','fun': lambda x: np.sum(x)-1},
            {'type':'eq','fun': lambda x: weighted_return(d_r,x)-r}
        )
        result = minimize(f,x0=[1/(d_r.shape[1]-2)]*(d_r.shape[1]-2),method='SLSQP',bounds=bounds,constraints=cons)
        res.append(result['fun'])
        w.append(result['x'])
    return res,w
```

```python
from scipy.optimize import minimize
return_level = np.arange(0.05,2,0.002)
vol_cml,wts_cml = eff_frontier(return_level,d_r_temp)
```

```python
from scipy.optimize import minimize
return_level = np.arange(0.05,2,0.002)
vol,wts = eff_frontier(return_level,d_r)
```

```python
# The tangency portfolio and its weights
eff_sharp = (return_level-rf)/vol
best_r = return_level[np.where(eff_sharp==np.max(eff_sharp))]
best_vol = np.array(vol)[np.where(eff_sharp==np.max(eff_sharp))]
best_wts_cml = wts_cml[np.where(eff_sharp==np.max(eff_sharp))[0][0]]
best_wts = wts[np.where(eff_sharp==np.max(eff_sharp))[0][0]]
best_r,best_vol,best_wts_cml,best_wts
```

## Re-balancing strategy : super optional

```python
year_d_r = {}
for year in range(2012,2021):
    year_data = data[(data['Date']<=str(year+1))&(data['Date']>str(year))]
    year_d_r[str(year)] = daily_return(year_data)
```

```python
def optimal_weights(return_level,d_r):
    vol,wts = eff_frontier(return_level,d_r)
    eff_sharp = (return_level-rf)/vol
    best_wts = wts[np.where(eff_sharp==np.max(eff_sharp))[0][0]]
    return best_wts
```

```python
# Note: Rerunning this could take a long time.
#       If you cannot see the image 'The performance of re-balance strategy (Started with the NAV=1)',
#       please see the folder. I have saved it as jpg.
optimal_weights_list = []
return_level = np.arange(0.05,2,0.002)
for year in range(2012,2020):
    y_d_r = year_d_r[str(year)]
    y_b_w = optimal_weights(return_level,y_d_r)
    optimal_weights_list.append(y_b_w)
    print(year)
```

```python
optimal_weights_list # The best weights for 2013-2020 and we assume the 2012 has the equal weights
```

```python
yearly_r = [((year_d_r['2012'].iloc[:,1:-1] @ np.array([1/8]*8).T)+1).cumprod().tolist()[-1]]
daily_r = ((year_d_r['2012'].iloc[:,1:-1] @ np.array([1/8]*8).T)+1).tolist()
for year in range(2013,2021):
    y_r = ((year_d_r[str(year)].iloc[:,1:-1] @ optimal_weights_list[year-2013].T)+1).cumprod().tolist()[-1]
    yearly_r.append(y_r)
    daily_r.extend(((year_d_r[str(year)].iloc[:,1:-1] @ optimal_weights_list[year-2013].T)+1).tolist())
```

```python
yearly_r # The yearly return+1
```

```python
# The total return in 2012-2020 (calculated by yearly returns)
rb_r = 1
for rr in yearly_r:
    rb_r *= rr
print('The total return in 2012-2020 is {}%'.format(round((rb_r-1)*100,2)))
```

```python
# Double check the calculation by daily returns
rb_r = 1
for rr in daily_r:
    rb_r *= rr
print('The total return in 2012-2020 is {}%'.format(round((rb_r-1)*100,2)))
```

```python
df_rb_dr = pd.DataFrame(columns=['Date'])
df_rb_yr = []
for item in year_d_r.values():
    df_rb_dr = pd.concat([df_rb_dr,item[['Date']]])
    df_rb_yr.append(item['Date'][item.shape[0]-1])
    
df_rb_dr = df_rb_dr.reset_index().drop(['index'],axis=1)
df_rb_dr = pd.concat([df_rb_dr,pd.DataFrame(daily_r,columns=['p'])],axis=1)
df_rb_dr['p'] = df_rb_dr['p'].cumprod()

df_rb_yr = pd.DataFrame(df_rb_yr,columns=['Date'])
df_rb_yr = pd.concat([df_rb_yr,pd.DataFrame(yearly_r,columns=['p'])],axis=1)
df_rb_yr['p'] = df_rb_yr['p'].cumprod()
```
